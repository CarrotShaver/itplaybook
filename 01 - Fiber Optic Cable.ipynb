### Fiber Optic Cables
- Single vs. Multi mode fiber
    - **Single Mode**
        - Cut-off Wavelength: Wavelength above which the fiber will only support a single mode
            - Cut-offs which are below the wavelength with support 2+ modes
            - Basically, lower cut-off wavelengths will support more modes than higher cut-off wavelengths
            - Cut-off wavelength is determined by the *Refractive Index Profile* -- The length and bend of the fiber               
        - Mode-field Diameter: Size of optical power distribution in fiber
            - Some optical power is distributed outside of the core within the cladding
                - For example 30% of the optical power of a fiber may come from the cladding
        - Typically used for long distances due to reliability of power, though it is costlier due to the small size
        - Yellow jacketed cables are generally single mode
        - Standard Diameters:
            - Core: 8-10um (micro-meters)
            - Cladding: 125um (micro-meters)
    - **Multi Mode**
        - Numerical Aperture: Measure of range of light accepted by fiber (angles of light)
        - Core Diameter: Core design of multi-mode fiber, where a larger core allows for more modes of light
            - Multi-mode has much larger core size than single-mode
        - Used mainly in LAN's due to the higher loss of light over long distances as light bounces within the larger diameter of the core
        - Larger size results in a cheaper cost, but a lower power
        - Orange jacketed cables are generally multi mode
        - Standard Diameters:
            - Core: 62.5um (micro-meters)
            - Cladding: 125um (micro-meters)

- Performance loss in fiber
    - Attenuation: Signal loss along length of fiber (light getting dimmer)
    - Dispersion: Distortion of a signal
    - Macro/Micro-bending: 
        - Macro: caused by attenutation over long lengths
        - Micro: caused by small bends and wraps made when installing fiber along corners, edges, etc.
- How it works: Light refraction off the cladding along the core
- Reliability/Durability: does not corrode or rust, is not vulnerable to electrical interference

- **Types of Fiber Modes**
    - SX and LX Multi-mode: 
        - Mainly used in LAN environments
        - Good up to ~550m
    - LX Single-mode: 
        - Good up to ~5km
        - Uses 1270-1335nm wavelength
    - LX10 Single-mode: 
        - Good up to ~10km
        - Uses 1360nm wavelength
    - EX Single-mode:
        - Good up to ~40km
        - Uses 1310nm wavelength
    - ZX Single-mode:
        - Good up to ~70km
        - Uses 1550nm wavelength
    - BX10 Single-mode:
        - Single Strand (Can do upstream and downstream with a single fiber)
            - Also known as "Bi-Di" (bi-directional)
            - Uses a pair of different wavelengths, one for downstream and one for upstream
        - 1480-1500nm downstream, 1260-1360 upstream
        - Good up to ~10km

- **Variables which affect Coupling**
    - End gap: small gap or space between fiber optic ends
    - Mismatch: ends are slightly above/below resulting in a loss of light
    - Dirt: Particles which are blocking light on a fiber cable
    - Axial Run-out: a bend on one end which may cause a small angle to be left open
    - Core mismatch: Two different sized core ends
